```python
data.corr()
```

* We have to perform correlation plot to check which of the factors are more related to the dependent variable(species) and which factors are less related to the dependent variable.

```python
#Correlation plot
fig =plt.figure(figsize=(10,7))
sns.heatmap(data.corr(),linewidths=1,annot=True)
```

* Sales and Profit are Moderately Correlated.
* Discount and Profit are Negatively Correlated
* Quantity and Profit are less Moderately Correlated

```python
plt.figure(figsize=(10,7))
sns.kdeplot(data['Sales'],color='red',label='Sales',shade=True,bw=25)
sns.kdeplot(data['Profit'],color='Blue',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

* Profit is more than that of sale but there are some areas where profit could be increased.

## Deal Analysis - State-wise :

```python
data['State'].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='State',data=data,palette='rainbow',order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.ylabel('Frequency / Number of deals')
plt.xlabel('States')
plt.title('State Wise Dealings', fontsize = 20)
plt.show()
```

* Top 3 states having highest deals are as follows:
    * California 
    * New York
    * Texas
* States having lowest deals are as follows :
    * Wyoming
    * West Virginia

## Deal Analysis - City-wise :

## State wise analysis of Profit, Discount, Sale

```python
data['State'].value_counts().head(10)
```

```python
df_state= data.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

### (i) State wise profit analysis

```python
df_state1=df_state.sort_values('Profit')
df_state1[['Profit']].plot(kind = 'bar', figsize = (15,4))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
```

* Highest Profit : Vermont
* Lowest Profit : Ohio

### (ii) State wise sales analysis :

```python
df_state['Sales'].plot(kind='pie', figsize = (20,20),autopct='%1.1f%%',startangle=90, shadow=True)
plt.title('State wise analysis of Sale',fontsize=20)
```

* Highest amount of sales= Wyoming(11.8%)
* Lowest amount of sales= South Dakota(0.8%)

### (iii) State wise Discount analysis :

```python
df_state1['Discount'].plot(kind='bar',figsize=(18,5))
plt.title('State wise analysis of Discount', fontsize=20)
```

* Illinois is the highest.

## City wise analysis of Profit 

```python
df_city= data.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head()
```

```python
#1.Low Profit
df_city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color = 'Pink')
plt.title('City wise analysis of Sale, Discount, profit')
```

## Quantity wise sales, Discount, Profit analysis :

```python
df_quantity = data.groupby(['Quantity'])[['Sales', 'Discount', 'Profit']].mean()
df_quantity.head(10)
```

```python
#1. sales 2. Discount 3. Profit
df_quantity.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(20, 20),
                     pctdistance=0.69,
                    startangle=90,     
                    shadow=True,
                    labels = df_quantity.index)
plt.title('Quantity wise analysis of Sale, Discount, profit')
```

## Sub-Category wise Sales, Profit and Discount

```python
df_sub_category = data.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

```python
#Based on sales 

plt.figure(figsize = (12,12))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
```

* Copier then Machine have High sales.

```python
#Based on Discount

plt.figure(figsize = (13,13))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
```

* Binders , Machines and then tables have high Discount.

```python
#Based on Profit

df_sub_category.sort_values('Profit')[['Sales','Profit']].plot(kind='bar', figsize= (10,5),label=['Avg Sales Price($)','Profit($)'])
```

* Copier: High Profit & sales
